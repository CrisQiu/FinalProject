# SE3: Model train, validate, and interpret RIC-SAT relationship

This notebook runs the SE3 workflow for the daily statistical relationship between:

- **RIC**: river ice concentration
- **SAT**: surface air temperature

## Default input directories

The input data come from the output of SE2 (daily RIC image) and added daily SAT derived from ERA5 archives:

- `FinalProject/raw_data/ERA5_temperature`
- `FinalProject/analysis_ready/RIC`

## Default output directory

- `FinalProject/Model_application`

The notebook calls the companion Python script and writes all tables and figures automatically.

```python
from pathlib import Path
import runpy

SCRIPT_PATH = Path(r"SE3_RIC_SAT_model_train_validate_interpret.py")
SCRIPT_PATH
```

## Run the full SE3 workflow

Run the next cell once. It will:

1. extract daily mean SAT from GeoTIFFs  
2. extract daily mean RIC from GeoTIFFs  
3. pair SAT and RIC by date  
4. fit linear and logistic models  
5. save summary tables and figures to `FinalProject/Model_application`

```python
runpy.run_path(str(SCRIPT_PATH), run_name="__main__")
```

## Inspect the main outputs

```python
from pathlib import Path
import pandas as pd

def find_project_root():
    cwd = Path.cwd().resolve()
    for base in [cwd, *cwd.parents]:
        if base.name == "FinalProject":
            return base
        candidate = base / "FinalProject"
        if candidate.exists():
            return candidate.resolve()
    return (cwd / "FinalProject").resolve()

PROJECT_ROOT = find_project_root()
OUTPUT_DIR = PROJECT_ROOT / "Model_application"
TABLE_DIR = OUTPUT_DIR / "tables"
FIGURE_DIR = OUTPUT_DIR / "figures"

print("Project root:", PROJECT_ROOT)
print("Output directory:", OUTPUT_DIR)
print("\nTables:")
for fp in sorted(TABLE_DIR.glob("*")):
    print(" -", fp.name)

print("\nFigures:")
for fp in sorted(FIGURE_DIR.glob("*")):
    print(" -", fp.name)
```

```python
summary_csv = TABLE_DIR / "model_summary.csv"
paired_csv = TABLE_DIR / "paired_daily_sat_ric.csv"

summary_df = pd.read_csv(summary_csv)
paired_df = pd.read_csv(paired_csv)

display(summary_df)
display(paired_df.head())
```
